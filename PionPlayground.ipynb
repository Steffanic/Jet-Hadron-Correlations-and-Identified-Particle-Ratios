```python
import ROOT 
import numpy as np
import matplotlib.pyplot as plt
import os
# load the pp file
f = ROOT.TFile("/mnt/d/pp/17p.root")
# get the tree
emcalList = f.Get("AliAnalysisTaskJetH_tracks_caloClusters_biased")
# get the histogram
JH = emcalList.FindObject("fhnJH")
# JH is a ThNsparse with 10 dimensions. Let's get axis 8 and setrangeuser to +-2 

# get the axis
JH.GetAxis(8).SetRangeUser(-2,2)
# then project the histogram to get a TH3D with 3 dimensions    and axes 3,4,6
th3d = JH.Projection(3,4,6)
# now i want to loop through each bin in thex and y dimension and plot the z projection
# first get the x and y axis
for i in range(1,th3d.GetNbinsX()+1):
    for j in range(1,th3d.GetNbinsY()+1):
        # get the projection
        th1d = th3d.ProjectionZ("th1d",i,i,j,j)
        nbins = th1d.GetNbinsX()
        plt.plot(np.linspace(-2,2,nbins),[th1d.GetBinContent(k) for k in range(1,nbins+1)])
        plt.show()
        plt.clf()
        

```
